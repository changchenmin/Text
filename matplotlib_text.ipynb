散佈圖

```python
# 散佈圖:plt.plot(x, y, 'o')

# 固定亂數種子值
random.seed(0)
# x軸的資料
x = np.random.randn(30)
# y軸的資料
y = np.sin(x) + np.random.randn(30)
# 指定圖形的大小(請試著修改20或6)
plt.figure(figsize=(20, 6))
# 描繪圖形
plt.plot(x, y, 'o')
# 也能如下描繪散佈圖
# plt.scatter(x, y)
# 標題
plt.title('Title Name')
# X軸標籤
plt.xlabel('X')
# Y軸標籤
plt.ylabel('Y')
# 顯示grid(圖形中的格線)
plt.grid(True)
```

```python
# 連續曲線:給予連續的值

# 固定亂數種子值
np.random.seed(0)
# 資料的範圍
numpy_data_x = np.arange(1000)
# 生成亂數與累積和
numpy_random_data_y = np.random.randn(1000).cumsum()
# 指定圖形大小
plt.figure(figsize=(20, 6))
# 可藉由label=與legend來加上曲線標籤
plt.plot(numpy_data_x, numpy_random_data_y, label='Label')
plt.legend()

plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
```

圖形的分割

```python
# 指定圖形大小
plt.figure(figsize=(20, 6))

# 2row 1column圖形的第一個
plt.subplot(2,1,1)
# numpy功能:.linspace，在一個範圍內，均勻撒點
x = np.linspace(-10, 10,100)
plt.plot(x, np.sin(x))

# 2row 1column圖形的第二個
plt.subplot(2,1,2)
# numpy功能:.linspace，在一個範圍內，均勻撒點
y = np.linspace(-10, 10,100)
plt.plot(y, np.sin(2*y))

plt.grid(True)
```

函數圖形的描繪

```python
def my_function(x):
    return x ** 2 + 2 * x + 1

# 產生-10到10的陣列
x = np.arange(-10, 10)
plt.figure(figsize = (20, 6))
plt.plot(x, my_function(x))
plt.grid(True)
```

直方圖

```python
# 固定亂數種子值
random.seed(0)
# 指定圖形大小
plt.figure(figsize = (20, 6))
# 描繪直方圖:plt.hist
plt.hist(np.random.randn(10 ** 5) * 10 + 50, bins = 60, range = (20, 80))
# plt.hist(arr, bins=10, range=(min, mix)
# arr:直方圖的一為陣列資料
# bins:直方圖的柱數，默認為10
# range:篩選數據範圍，默認是最小到最大
plt.grid(True)
```

練習三
產生兩組從0到1之間均勻亂數1000個，分別描繪直方圖。
但為了將各自的直方圖分開顯示，請使用plt.subplot。這裡所謂的均勻亂數，是指從某數到另一
數之間等機率產生的亂數，使用np.random.uniform。比如說，要產生10個從0到1之間的數字
時，寫成np.random.uniform(0.0, 1.0, 10)。
此外，除了1000個，也試試100個與10,000個吧。是否有什麼發現呢?

```python
def uni_hist(N):

    x = np.random.uniform(0.0, 1.0, N)
    y = np.random.uniform(0.0, 1.0, N)

    plt.subplot(2, 1, 1)
    plt.hist(x)
    plt.title("No1:histogram")

    plt.subplot(2, 1, 2)
    plt.hist(y)
    plt.title("No2:histogram")

    plt.grid(True)
    plt.tight_layout()

uni_hist(1000)
```
